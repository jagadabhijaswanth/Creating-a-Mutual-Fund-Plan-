```python
fig = go.Figure()

for company in data.columns[1:]:
    fig.add_trace(
        go.Scatter(
            x= data['Date'],
            y= data[company],
            mode= 'lines',
            name= company
        )
    )

fig.update_layout(
    title = 'stock price trends of companies',
    xaxis_title = 'Date',
    yaxis_title = 'closing price(rupees)',
    xaxis = dict(
        tickangle = 45
    ),
    legend=dict(
        x=1.05,
        y=1,
        traceorder="normal",
        font=dict(size=10),
        orientation="v"
    ),
)
fig.show()
```

companies with the highest risks for investing


```python
companies = data.columns[1:]
volatality_of_comapnies = data[companies].std()
volatality_of_comapnies.sort_values(ascending=False).head(10)
high_risk_comapanies = volatality_of_comapnies.sort_values(ascending=False).head(10)


fig = go.Figure()

fig.add_trace(
    go.Bar(
        x = high_risk_comapanies.index,
        y = high_risk_comapanies.values 
    )
)

fig.update_layout(
    title = 'high_risk_comapanies',
    xaxis_title = 'companies',
    yaxis_title = 'volatality'
)
```

```python
companies = data.columns[1:]
growth_of_all_companies = data[companies].pct_change()*100
avg_growth_of_companies = data[companies].mean()
avg_growth_of_companies.sort_values(ascending=False).head(10)
```

companies with the highest return on investments

```python
initial_prices = data[companies].iloc[0]
final_prices = data[companies].iloc[-1]
highest_return_percentage = ((final_prices - initial_prices)/initial_prices)*100
highest_return_percentage.sort_values(ascending=False).head(10)
```
